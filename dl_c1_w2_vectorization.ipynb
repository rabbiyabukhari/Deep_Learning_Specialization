```python
import numpy as np

a=np.array([1,2,3,4])
print(a) #array created
```

```python
import time

a=np.random.rand(1000000)
b=np.random.rand(1000000)
tic=time.time()
c=np.dot(a,b)
toc=time.time()
print(c)
print("Vectorized version: "+ str(1000*(toc-tic))+" ms"+"\n") # time taken for vectorized version code


c=0
tic=time.time()
for i in range(1000000):
    c+=a[i]*b[i]
toc=time.time()
print(c)
print("Non-Vectorized Version(For loop): "+ str(1000*(toc-tic))+" ms") # time taken for non-vectorized version(for loop) code

#vectorized version runs about 300 times faster
```

Both GPU and CPU have parallelization instruction called SIMD(single instruction multiple data)- means use builtin functions or others without explicit for loops, it takes python numpy to have better advantage of parallelism.
GPUs is remarkably good with SIMD calculations, but CPU isn't bad either.
